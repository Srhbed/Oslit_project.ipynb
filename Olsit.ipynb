```python
from sklearn.compose import make_column_transformer, make_column_selector, ColumnTransformer
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, OneHotEncoder, RobustScaler
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.cluster import KMeans

one_hot_encoder = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(), data_categ)]
    
)
```

```python
scaler=StandardScaler()
data_norm=pd.DataFrame(scaler.fit_transform(data_num),columns=data_num.columns)
```

```python
data_norm
```

```python
tab= []
for i in range (1, 11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit (data_norm)
    tab.append(kmeans.inertia_)
plt.plot (range (1,11), tab)
plt.title ("La méthode Eblow")
plt.xlabel("nombre de clusters")
plt.ylabel("Inertie")
```

```python
from sklearn.decomposition import PCA
# from sklearn.cluster import KPrototypes


# Effectuer une ACP sur les résultats
pca = PCA(n_components=2)
X_pca = pca.fit_transform(data_norm)


plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data_norm["cluster"])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Visualisation des clusters obtenus par Kprototypes avec ACP')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
cluster=kmeans.fit_predict (data_norm)
data_norm['cluster']=cluster
```
